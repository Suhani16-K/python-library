### Pandas Series

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

### Importing Pandas

```python
import numpy as np
import pandas as pd
```

### Series from lists

```python
#String
country = ['India','Pakistan','USA','Nepal','Srilanka']
pd.Series(country)
```

```python
# integers
runs=[43,55,65,76,87,89,90]
pd.Series(runs)
```

```python
# custom index
marks=[43,54,56,45]
subjects = ['maths','english','science','hindi']

pd.Series(marks,index=subjects)
```

```python
# setting a name
marks=pd.Series(marks,index=subjects,name='Suhani ke marks')
marks
```

### Series from dict

```python
marks  = {'maths':67,'english':65,'science':87,'hindi':78}
ms=pd.Series(marks,name='suhani ke number')
ms
```
